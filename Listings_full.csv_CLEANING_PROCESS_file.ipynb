# Below is a first-pass cleaning of "listings_full.csv"

```python
import pandas as pd
```

```python
# import csv file
listings_full_df = pd.read_csv("./Resources/listings_full.csv")
pd.set_option('display.max_columns', None)
listings_full_df.head()
```

```python
listings_full_df.shape
```

# NOTE, this drops the columns from the "Airbnb Data Deep Dive"'s "Fields not required" tab 

# AND "id"

```python
listings_full_df = listings_full_df.drop(["id", "neighbourhood", "notes", "transit", "access", "interaction", "house_rules", "thumbnail_url", "medium_url", "picture_url", "xl_picture_url", "host_id", "host_url", "host_name", "host_acceptance_rate", "host_thumbnail_url", "host_picture_url", "host_neighbourhood", "host_verifications", "neighbourhood", "smart_location", "country_code", "country", "state", "is_location_exact", "calendar_updated", "requires_license", "license", "jurisdiction_names"], axis=1)
listings_full_df.head()
```

```python
listings_full_df.shape
```

# Determine NaN counts for each column using a generator function:

```python
[*listings_full_df.isnull().sum().sort_values().items()]
```

## Now, since "neighbourhood_group_cleansed", "weekly_price", "monthly_price", and "square feet" are all about 99% NaNs we will need to drop these columns as well:

```python
listings_full_df = listings_full_df.drop(columns=["neighbourhood_group_cleansed","weekly_price", "monthly_price", "square_feet"])
```

```python
[*listings_full_df.isnull().sum().sort_values().items()]
```

# Since we've dropped the columns with 99% NaNs, we now have the ability to drop rows of missing data without losing so much of our data.  

# However, now, we have the option/decision for whether to drop more columns that still have lots of NaN's in them first or to go ahead and drop all the rows with missing values (NaNs).  Of course, dropping NaNs here is a trade-off without a perfect solution, but now if we drop the rows with NaNs we are still left with nearly 2,000 useful rows from which to work.

```python
listings_full_df = listings_full_df.dropna()
listings_full_df
```
