```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Resources/listings_full.csv")
```

```python
data.head()
```

```python
#  List out all the variables in the dataset
data.columns.values
```

```python
# do not keep all the descriptive variables
# do not keep all photo variables
# do not keep all links
# do not keep all host name and description variables
# host_listings_count and host_total_listings_count are redundant drop the host_total
# do not keep street - doesn't make sense just shows city and state
# do not keep Amenities - using parsed file
# do not keep calculated fields
# we are keeping lat/long and neighborhood cleansed so don't need other location identifiers
# market is redundant as we are only looking at nashville market - should be kept if doing multi markets
# do not need country see above

```

```python
#Create new dataframe with only selected variables

data_new = data[['id','last_scraped','host_since','host_listings_count','host_is_superhost','host_identity_verified','neighbourhood_cleansed',\
'latitude','longitude','room_type','property_type','accommodates','bathrooms','bedrooms','beds','bed_type',\
'square_feet','price','weekly_price','monthly_price','security_deposit','cleaning_fee','review_scores_rating',
'number_of_reviews','review_scores_cleanliness',\
'review_scores_location','review_scores_communication','review_scores_checkin','review_scores_value','instant_bookable',\
'is_business_travel_ready','cancellation_policy','require_guest_profile_picture','require_guest_phone_verification',\
'has_availability' ]]        

```

```python
pd.set_option('display.max_columns', None)
data_new.head()
```

```python
#  check to see if any columns are mostly nan
[*data_new.isnull().sum().sort_values().items()]
```

```python
# drop the columns with mostly NaN
data_new=data_new.drop(columns=['weekly_price','monthly_price','square_feet'])
data_new.head()
```

```python
#  the following sections change the string $ fields to float
```

```python
data_new['price']=data_new['price'].str.replace('$','')
data_new['price']=data_new['price'].str.replace(',','')
data_new['price']=data_new['price'].astype(float)
```

```python
data_new['security_deposit']=data_new['security_deposit'].str.replace('$','')
data_new['security_deposit']=data_new['security_deposit'].str.replace(',','')
data_new['security_deposit']=data_new['security_deposit'].astype(float)
data_new.head()
```

```python
data_new['cleaning_fee']=data_new['cleaning_fee'].str.replace('$','')
data_new['cleaning_fee']=data_new['cleaning_fee'].str.replace(',','')
data_new['cleaning_fee']=data_new['cleaning_fee'].astype(float)
data_new.head()
```

```python
data_new.dtypes
```

```python
# the next section resets NaNs to $0 for cleaning dep and security dep
```

```python
[*data_new.isnull().sum().sort_values().items()]
```

```python
data_new['security_deposit']=data_new['security_deposit'].fillna(0)
```

```python
data_new['cleaning_fee']=data_new['cleaning_fee'].fillna(0)
```

```python
# drop 'beds'  many have 0s and some nans so not descriptive
data_new=data_new.drop(columns=['beds'])
```

```python
data_new.dtypes
```

```python
[*data_new.isnull().sum().sort_values().items()]
```

```python
# create new variable yes or no as to reviews so if this is no and the review is 0
# reviews range from 2 to 10 there are no real 0 reviews
data_new['is_review'] = np.where(data_new['review_scores_rating']>0,True,False)
data_new[:50]


```

```python
#  replace all NaNs in review columns with 0 - note there are no real 0 reviews and there is a variable is_review to identify the fake 0 values

cols=['review_scores_value','review_scores_location','review_scores_checkin','review_scores_communication','review_scores_cleanliness','review_scores_rating']
data_new[cols]=data_new[cols].fillna(0)
```

```python
data_new[:50]
```

```python
# Calculate number of days hosted
import datetime
date1=data_new['last_scraped'][0]
date2=data_new['host_since'][0]
mdate1 = datetime.datetime.strptime(date1, "%Y-%m-%d").date()
rdate1 = datetime.datetime.strptime(date2, "%Y-%m-%d").date()
delta =  (mdate1 - rdate1).days
delta

```

```python
# add new variable called "days_host" which is a calculated value of the difference between the scrape date and the host_since date

data_new[['last_scraped','host_since']] = data_new[['last_scraped','host_since']].apply(pd.to_datetime) #if conversion required
data_new['days_host'] = (data_new['last_scraped'] - data_new['host_since']).dt.days

```
